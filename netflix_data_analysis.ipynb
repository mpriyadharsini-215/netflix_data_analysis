```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Priya\Downloads\archive\netflix_titles.csv")


# =========================
# 1. Movies vs TV Shows Trend
# =========================
print("\n MOVIES AND TV SHOWS OVER TIME \n")
grp_yr = df.groupby(['release_year', 'type']).size().unstack()
print(grp_yr)
grp_yr.plot(kind='line')
plt.title("MOVIES AND TV SHOWS OVER TIME")
plt.xlabel("Year")
plt.ylabel("Content Count")
plt.show()
print("\n The data shows that Netflix content experienced rapid growth between 2015 and 2018, reaching its peak during this period. After 2018, growth slowed down, with movies slightly declining while TV shows continued to grow steadily.\n")

# =========================
# 2. Country Analysis
# =========================
df['country'] = df['country'].str.split(',')
df = df.explode('country')
grp_country = df.groupby(['release_year', 'country']).size().unstack()
recent = grp_country.loc[2018:]
growth = recent.diff().mean().sort_values(ascending=False).head(5)
print("\n TOP 5 COUNTRIES - RECENT YEAR GROWTH \n")
print(growth)
growth.plot(kind='bar')
plt.title("TOP 5 COUNTRIES - CONTENT GROWTH")
plt.xlabel("Country")
plt.ylabel("Average Growth per Year")
plt.show()
print("\n The analysis shows that countries like Kuwait, Taiwan, and Romania have the highest recent growth rates, indicating Netflix's expansion into emerging markets and increasing global diversification.\n")

# =========================
# 3. Genre Analysis
# =========================
df['listed_in'] = df['listed_in'].str.split(',')
df = df.explode('listed_in')
count_genre = df['listed_in'].value_counts().head(5)
print("\n TOP 5 GENRES ON NETFLIX \n")
print(count_genre)
count_genre.plot(kind='bar')
plt.title("TOP 5 GENRES ON NETFLIX")
plt.xlabel("Genre")
plt.ylabel("Content Count")
plt.show()
print("\n International Movies and Drama dominate Netflix, indicating a strong focus on globally appealing and emotionally engaging content.\n")

# =========================
# 4. Ratings Analysis
# =========================
print("\n TOP 5 CONTENT RATINGS ON NETFLIX \n")
count_rating = df['rating'].value_counts().head(5)
print(count_rating)
count_rating.plot(kind='bar')
plt.title("TOP 5 CONTENT RATINGS ON NETFLIX")
plt.xlabel("Rating")
plt.ylabel("Content Count")
plt.show()
print("\n Most content falls under TV-MA and TV-14, suggesting Netflix primarily targets mature audiences.\n")

# =========================
# FINAL CONCLUSION
# =========================
print("\n FINAL INSIGHT \n")

print("Netflix shows strong global expansion, with rapid growth between 2015–2018 followed by stabilization.")
print("The platform focuses heavily on international content, diverse genres, and mature audiences.")
print("Overall, Netflix's strategy is driven by global reach and long-term user engagement through varied content offerings.\n")
```
